```python
#1. Introduction
# Introduction
"""
This project analyzes product sales data from Amazon to understand sales trends,
popular products, and customer preferences. 
"""

```

```python
#2. Data Import and Exploration
import pandas as pd
import numpy as np

# Load the dataset
url = 'cleaned.csv'
dataset = pd.read_csv(url)

# Display the first few rows of the dataset
print(dataset.head())
```

```python
#3. Data Cleaning
# Check for missing values
print(dataset.isnull().sum())

# Display data types
print(dataset.dtypes)
```

```python
#4. Exploratory Data Analysis (EDA)
#a. Summary Statistics
# Summary statistics of numerical columns
print(dataset.describe())
```

```python
#b. Unique Values Count
# Count unique users and items
print(f"Unique Users: {dataset['user_id'].nunique()}")
print(f"Unique Items: {dataset['item_id'].nunique()}")
```

```python
#5. Visualizations
#a. Yearly Sales Analysis
import matplotlib.pyplot as plt

# Yearly sales analysis
yearly_sales = dataset.groupby('year')['amount'].sum()
yearly_sales.plot(kind='bar', title='Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
#b. Monthly Sales Analysis (2015-2018)
monthly_sales = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)].groupby('month')['amount'].sum()
monthly_sales.plot(kind='bar', title='Monthly Sales (2015-2018)')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
#c. Top Selling Brands (2015-2018)
top_brands = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)].groupby('brand')['amount'].sum().sort_values(ascending=False).head(10)
top_brands.plot(kind='bar', title='Top Selling Brands (2015-2018)')
plt.xlabel('Brand')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
#6. Conclusion
# Conclusion
"""
The analysis revealed that:
- The highest sales occurred in 2015.
- January was consistently the best month for sales.
- Bose emerged as the top-selling brand during the analyzed period.
"""
```
